```python
import networkx as nx
import random

def generate_erdos_renyi_graph(number_of_nodes, number_of_edges, random_seed):
    random.seed(random_seed)
    graph = nx.erdos_renyi_graph(number_of_nodes, round(number_of_edges / (number_of_nodes * (number_of_nodes - 1) / 2)), directed=False)
    return graph

if __name__ == "__main__":
    graph = generate_erdos_renyi_graph(10000, 100000, 10)
    print(graph.number_of_nodes())
    print(graph.number_of_edges())
```
